```python
#collection Mnipulator1

print("welcome to the student data collector ")
student_records=[]
while True:
    print("1.new student add:\n2.display the information for all student\n3.update studenet information\n4.delete student\n5.display subject offerd\n6.exit")
    print("=*=*=*=*=*=*=*=*=*=*=*=*=*=*=*=*=*=*=*=*=*=*=*=")
    choose = int(input("enter your choice:"))
    match choose:
        case 1:
            student_id=int(input("enter student id:"))
            student_name=str(input("enter student name:"))
            student_age=int(input("enter student age:"))
            student_grade=input("enter student grade:")
            student_dob=input("enter date of birth(dd-mm-yyyy)")
            student_sub=str(input("enter student's subject:"))

            student = {
                      "student_id":student_id,
                      "student_name":student_name,
                      "student_age":student_age,
                      "student_grade":student_grade,
                      "student_dob":student_dob,
                      "student_sub": student_sub
                      
                       
            }
            student_records.append(student)
            print("your record has successfully added")
        case 2:
            if len(student_records)==0:
                print("Not record found")
            else:
                print("+-+-+-+-+-+-+STUDENT RECORDS+-+-+-+-+-+-+-")
                for student in student_records:
                    print( "student_id:",student["student_id"],end="")
                    print("student_name:",student["student_name"],end="|")
                    print("student_age:",student["student_age"],end="|")
                    print("student_grade:",student["student_grade"],end="|")
                    print("student_dob:",student["student_dob"],end="|")
                    print("student_sub:",student["student_sub"],end="|")
                    print("\n")
               
        case 3:
           if len(student_records) == 0:
                print("No student records found.")
           else:
                update_id = int(input("Enter Student ID to update: "))
                

                for student in student_records:
                    if student["student_id"] == update_id:
                        
                        student["student_name"] = input("Enter new student name: ")
                        student["student_age"] = int(input("Enter new student age: "))
                        student["student_grade"] = input("Enter new student grade: ")
                        student["student_dob"] = input("Enter new Date of Birth (dd-mm-yyyy): ")
                        student["student_sub"] = input("Enter new student subject: ")
                        print("Student record updated successfully.")
                        
        case 4:
             if len(student_records) == 0:
                print("No student records found.")
             else:
                delete_id = int(input("Enter Student ID to delete: "))
                for student in student_records:
                    if student["student_id"] == delete_id:
                        student_records.remove(student)
                        
                        print("Student record deleted successfully.")
                        break

                 
        case 5: 
            if len(student_records) == 0:
                print("No student record found")
            else:
                 print("Subjects Offered:")
                 subjects = []

                 for student in student_records:
                     if student["student_sub"] not in subjects:
                         subjects.append(student["student_sub"])

                 for subject in subjects:
                     print(subject)
        case 6:
            print("Thank You!!for using collection manipulator")
            break
        case _:
            print("invalid choice")
```
